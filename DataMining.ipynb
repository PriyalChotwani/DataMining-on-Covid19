```python
#Reading Hourly Calories Data Set
hourly_cal = pd.read_csv(r"C:/Users/91965/Desktop/dataset/Fitabase Data 4.12.16-5.12.16/hourlyCalories_merged.csv")
hourly_cal.head()
```

```python
#To display the information of the table
hourly_cal.info()

#Display additional Information
print('\n Number of duplicate entries =',hourly_cal.duplicated().sum())
print(' Total Entries               =',hourly_cal.Id.nunique())

#Describing the information
hourly_cal.describe()
```

```python
#Converting ActivityHour Column into useful format
hourly_cal.ActivityHour = pd.to_datetime(hourly_cal.ActivityHour)

#Converting into String
hourly_cal.ActivityHour = hourly_cal.ActivityHour.astype(str)

#Split the ActivityHour string into Date and Time Columns
hourly_cal["Date"] = hourly_cal.ActivityHour.str.split(expand=True)[0]
hourly_cal["Time"] = hourly_cal.ActivityHour.str.split(expand=True)[1]

#Formatting Date
hourly_cal.Date = pd.to_datetime(hourly_cal.Date)

#Identfying Day from Date Column
hourly_cal["Day"] = hourly_cal.Date.dt.day_name()

#Display the final dataset
print("Final Hourly Calories Dataset:")
hourly_cal.head()
```

```python
#Reading Daily Sleep DataSet
daily_sleep = pd.read_csv(r"C:/Users/91965/Desktop/dataset/Fitabase Data 4.12.16-5.12.16/sleepDay_merged.csv")
daily_sleep.head()
```

```python
#Cleaning the dataset: Daily Activity
daily_sleep = daily_sleep[["Id","SleepDay","TotalMinutesAsleep"]]
```

```python
#To display the information of the table
daily_sleep.info()

#Display additional Information
print('\n Number of duplicate entries =',daily_sleep.duplicated().sum())
print(' Total Entries               =',daily_sleep.Id.nunique())

#Describing the information
daily_sleep.describe()
```

```python
#Converting SleepDay Column into useful format
daily_sleep.SleepDay = pd.to_datetime(daily_sleep.SleepDay)

#Converting into String
daily_sleep.SleepDay = daily_sleep.SleepDay.astype(str)

#Split the SleepDay string into Date and Time Columns
daily_sleep["Date"] = daily_sleep.SleepDay.str.split(expand=True)[0]

#Formatting Date
daily_sleep.Date = pd.to_datetime(daily_sleep.Date)

#Identfying Day from Date Column
daily_sleep["Day"] = daily_sleep.Date.dt.day_name()

#Remove Duplicate Entries
daily_sleep.drop_duplicates(inplace = True)

#Display the final dataset
print("Final Daily Sleep Dataset:")
daily_sleep.head()
```
